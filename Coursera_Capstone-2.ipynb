```python
import pandas as pd
import numpy as np
import requests 
import random 

from geopy.geocoders import Nominatim 
from IPython.display import Image 
from IPython.core.display import HTML 
from IPython.display import display_html

from pandas.io.json import json_normalize
import folium 
from bs4 import BeautifulSoup
from sklearn.cluster import KMeans
import matplotlib.cm as cm
import matplotlib.colors as colors

print('All libraries imported successfully')
```

```python
#Webscrapping contents from Wilkepedia

link = requests.get('https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M&oldid=1011037969').text
soup = BeautifulSoup(link,'lxml')
print(soup.title)
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

```python
#Convert web extract to panda DataFram
df = pd.read_html(tab)
df=df[0]
df.head()
```

```python
# Drop the rows where Borough is 'Not assigned'

df1 = df[df.Borough != 'Not assigned']

# Combining the neighbourhoods with Postalcode
df2 = df1.groupby(['Postal Code','Borough'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)

# Replacing the name of the neighbourhoods which are 'Not assigned' with names of Borough
df2['Neighbourhood'] = np.where(df2['Neighbourhood'] == 'Not assigned',df2['Borough'], df2['Neighbourhood'])

df2
```

```python
cordinates = pd.read_csv('https://cocl.us/Geospatial_data')
cordinates.head()
```

```python
#Rename column headers
cordinates.rename(columns={'Postcode':'Postal Code'},inplace=True)
```

```python
cordinates.columns
```

```python
#Merge both tables into one table
df3 = pd.merge(df2,cordinates, on = 'Postal Code')
df3.head()
```

```python
#Filter for Toronto

df4 = df3[df3['Borough'].str.contains('Toronto',regex=False)]
df4
```

```python
#Mapping the neighbourhoods

torontomap = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighbourhood in zip(df4['Latitude'],df4['Longitude'],df4['Borough'],df4['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(torontomap)
torontomap
```

```python
# Introduce K-Means to cluster the neighborhoods

k=10
c = df4.drop(['Postal Code','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(c)
kmeans.labels_
#df4.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
df4
```

```python
# mapping the clusters
map_cl = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
y = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(y)))
rainbow = [colors.rgb2hex(i) for i in colors_array]
```

```python
# mark to the map

mark_colors = []

for lat, lon, neighbourhood, cluster in zip(df4['Latitude'], df4['Longitude'], df4['Neighbourhood'], df4['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=6,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.8).add_to(map_cl)
       
map_cl
```
